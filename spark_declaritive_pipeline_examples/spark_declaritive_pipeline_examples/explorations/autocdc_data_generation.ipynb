### Create Dummy Requests data


```python
%python
spark.sql("""
CREATE OR REPLACE TABLE nethra.sdp.requests_raw (
  _id STRING,
  requestedBy STRING,
  customValues STRING,
  ingestion_time TIMESTAMP
)
TBLPROPERTIES (delta.enableChangeDataFeed = true)
""")

# Table is now empty and CDF is enabled

display(spark.sql("DESCRIBE EXTENDED nethra.sdp.requests_raw"))
```

```python
display(spark.sql("""
SELECT 
  version,
  DATE_FORMAT(timestamp, 'yyyy-MM-dd HH:mm:ss') AS timestamp,
  operation,
  userName,
  operationMetrics.numOutputRows AS output_rows,
  operationMetrics.numAddedFiles AS added_files,
  operationMetrics.numRemovedFiles AS removed_files,
  CAST(operationMetrics.executionTimeMs AS BIGINT) / 1000.0 AS execution_time_sec,
  operationParameters
FROM (DESCRIBE HISTORY nethra.sdp.requests_raw)
WHERE operation IN ('WRITE', 'DELETE', 'UPDATE', 'MERGE')
ORDER BY version DESC
"""))
```
